```python
import os
import re

folder = "data/new_data"  # path to your folder

# regex to match filenames like "Ğ—Ğ²Ñ–Ñ‚-1.18.xls"
pattern = re.compile(r"Ğ—Ğ²Ñ–Ñ‚-(\d+)\.(\d+)\.xls")

for filename in os.listdir(folder):
    match = pattern.match(filename)
    if match:
        month, year = match.groups()
        new_name = f"{year}-{month}.xls"
        old_path = os.path.join(folder, filename)
        new_path = os.path.join(folder, new_name)
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} â†’ {new_name}")
```

```python
import pandas as pd
from pathlib import Path

FOLDER = Path("data/new_data")         # folder with raw csvs
OUT_FOLDER = Path("data/new_data_cleaned")  # folder for cleaned csvs
OUT_FOLDER.mkdir(parents=True, exist_ok=True)  # create if missing


def convert_one(csv_path: Path):
    # 1ï¸âƒ£ Read CSV with 2 header rows
    df = pd.read_csv(csv_path, header=[0, 1])

    # 2ï¸âƒ£ Remove first and last data rows
    df = df.iloc[0:-1]

    # 3ï¸âƒ£ Drop completely empty rows
    df = df.dropna(how="all")

    # 4ï¸âƒ£ Remove columns that are all zeros
    df = df.loc[:, (df != 0).any(axis=0)]

    # 5ï¸âƒ£ Save cleaned file to new folder
    out_path = OUT_FOLDER / f"{csv_path.stem}_clean.csv"
    df.to_csv(out_path, index=False, encoding="utf-8-sig")

    # âœ… Use absolute() to avoid ValueError
    print(f"âœ” {csv_path.name} â†’ {out_path.absolute()}")


if __name__ == "__main__":
    for csv_file in FOLDER.glob("*.csv"):
        convert_one(csv_file)

    print("âœ… All done. Cleaned files saved in:", OUT_FOLDER.absolute())
```

```python
import pandas as pd
from pathlib import Path

FOLDER = Path("data/new_new_data_cleaned")           # folder with raw csvs
OUT_FOLDER = Path("data/clean_csv_2")  # folder for cleaned csvs
OUT_FOLDER.mkdir(parents=True, exist_ok=True)  # create if missing

# mapping of Ukrainian â†’ English column names
RENAME_MAP = {
    "Ğ’Ğ¸Ğ´ Ñ‚Ğ¾Ğ²Ğ°Ñ€Ñƒ": "category",
    "Ğ¡ÑƒĞ¼Ğ° Ğ¿Ğ¾ÑÑ‚Ğ°Ñ‡Ğ°Ğ»ÑŒĞ½Ğ¸ĞºĞ°": "supplier_sum",
    "Ğš-ÑÑ‚ÑŒ Ñ€ĞµĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ—": "sold_qty",
    "Ğ¡Ğ¾Ğ±Ñ–Ğ²Ğ°Ñ€Ñ‚Ñ–ÑÑ‚ÑŒ Ñ€ĞµĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ—": "sold_cost",
    "Ğ¡ÑƒĞ¼Ğ° Ñ€ĞµĞ°Ğ»Ñ–Ğ·Ğ°Ñ†Ñ–Ñ—": "sold_sum",
    "Ğš-ÑÑ‚ÑŒ Ğ¿Ğ¾Ğ²ĞµÑ€Ğ½ĞµĞ½Ğ½Ñ": "return_qty",
    "Ğ¡Ğ¾Ğ±Ñ–Ğ²Ğ°Ñ€Ñ‚Ñ–ÑÑ‚ÑŒ Ğ¿Ğ¾Ğ²ĞµÑ€Ğ½ĞµĞ½Ğ½Ñ": "return_cost",
    "Ğ¡ÑƒĞ¼Ğ° Ğ¿Ğ¾Ğ²ĞµÑ€Ğ½ĞµĞ½Ğ½Ñ": "return_sum",
    "Ğ’Ğ¸Ñ‚Ğ¾Ñ€Ğ³": "revenue",
    "Ğ’Ğ¸Ñ€ÑƒÑ‡ĞºĞ°": "profit",
}


def convert_one(csv_path: Path):
    try:
        # 1ï¸âƒ£ Read CSV â€” if it has 2 headers, skip the first one
        df = pd.read_csv(csv_path)
    except pd.errors.ParserError:
        # fallback: if no second header exists, read normally
        df = pd.read_csv(csv_path)

    # 2ï¸âƒ£ Remove first and last data rows (if exist)
    if len(df) > 2:
        df = df.iloc[0:-1]

    # 3ï¸âƒ£ Drop completely empty rows
    df = df.dropna(how="all")

    # 4ï¸âƒ£ Remove columns that are all zeros
    df = df.loc[:, (df != 0).any(axis=0)]

    # 5ï¸âƒ£ Rename columns if they match known names
    df = df.rename(columns={col: RENAME_MAP.get(col, col) for col in df.columns})

    # 6ï¸âƒ£ Save cleaned file
    out_path = OUT_FOLDER / f"{csv_path.stem}_clean.csv"
    df.to_csv(out_path, index=False, encoding="utf-8-sig")

    print(f"âœ” {csv_path.name} â†’ {out_path.absolute()}")


if __name__ == "__main__":
    for csv_file in FOLDER.glob("*.csv"):
        convert_one(csv_file)

    print("âœ… All done. Cleaned & renamed files saved in:", OUT_FOLDER.absolute())
```
